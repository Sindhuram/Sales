### Merging the two dataframes via left join

```python
df = pd.merge(store_data1, store_data2, how='left', on='Store', suffixes=('_left', '_right'))
```

```python
#This makes sure that all the data from store_data1 will be displayed,irrespective of whether the keys in the first
#Dataframe can be found in the second Dataframe. Whereas, for the second Dataframe, only the records with the keys in 
#the second Dataframe that can be found in the first Dataframe will be displayed.
```

### Splitting the data

```python
X = df.drop(['sales', 'date', 'competitionopensinceyear', 'promo2sinceweek','promo2sinceyear'], axis=1)
```

```python
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=1)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
fig, ax = plt.subplots(figsize=(15,10))
sns.heatmap(X_train.corr(), annot=True, cmap='YlGnBu')
```

### Loading the data

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
print(lr.intercept_)
```

```python
print(lr.coef_)
```

```python
X_train.columns
```

```python
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_train,y_train_pred))
print(r2_score(y_test,y_test_pred))
```
